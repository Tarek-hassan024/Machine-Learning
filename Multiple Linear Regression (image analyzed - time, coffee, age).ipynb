## Splitting the variables

```python
X = pd.DataFrame(df[['Coffee', 'Age', 'Time']])
y = pd.DataFrame(df.Images_Analyzed)
```

## Model

```python
lm = LinearRegression()
model = lm.fit(X,y)
```

## Prediction

```python
lm.predict([[1,22,10]])[0][0].round(2)
```

## Model Evaluation

```python
lm.score(X,y)
```
