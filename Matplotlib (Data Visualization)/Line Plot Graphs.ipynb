## Necessary Packages

```python
import pandas as pd
import pandas_datareader as pdr
import matplotlib.pyplot as plt
#from matplotlib import rcParams   #It's for customize the size of the graph
```

## Loading datasets

```python
stock = ["GOOG","AMZN"]
data = pdr.get_data_yahoo(stock, start = '2019-01-01')['Close']
data.head()
```

## Primary Line Plot

```python
plt.plot(data.AMZN, color = "Blue")
plt.plot(data.GOOG, color = "Green")
```

## Legend

```python
plt.plot(data.AMZN, label = "AMZN")
plt.plot(data.GOOG, label = "GOOG")
plt.legend()
```

## Location of the legend

```python
plt.plot(data.AMZN, label = "AMZN")
plt.plot(data.GOOG, label = "GOOG")
plt.legend(loc = 1)
```

## Gridline

```python
plt.plot(data.AMZN, label = "AMZN")
plt.plot(data.GOOG, label = "GOOG")
plt.legend(loc = 2)
plt.grid(True, color = "Red", linestyle = ":")
```

## Axis Labels

```python
plt.plot(data.AMZN, label = "AMZN")
plt.plot(data.GOOG, label = "GOOG")
plt.legend(loc = 2)
plt.xlabel("Year")
plt.ylabel("Stocks")
plt.title("Google vs Amazon")
```

## Axis Limits and ticks

```python
plt.plot(data.AMZN, label = "AMZN")
plt.plot(data.GOOG, label = "GOOG")
plt.legend(loc = 2)
plt.ylim(500,5000)
plt.yticks([500,2000,4000,5000])
```
